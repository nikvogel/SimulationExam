```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def temp_function(t, alpha, i):
    return t*(alpha**i)
```

```python
t_6 = np.exp(6/12)
t_12 = np.exp(12/12)
```

```python
iterations = np.arange(0,1000)
```

```python
temp_6 = []
temp_12 = []
```

```python
for i in iterations:
    temp_6.append(temp_function(t_6, 0.99, i))
    temp_12.append(temp_function(t_12, 0.99, i))
```

```python
plt.plot(iterations, temp_6, label = '6 periods')
plt.plot(iterations, temp_12, label = '12 periods')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Temperature')
plt.tight_layout()
plt.savefig('TempPlot.png')
```

```python
metropolis_6_1 = []
metropolis_6_10 = []
metropolis_12_1 = []
metropolis_12_10 = []
for i in iterations:
    metropolis_6_1.append(np.exp(-1 / temp_6[i]))
    metropolis_6_10.append(np.exp(-10 / temp_6[i]))
    metropolis_12_1.append(np.exp(-1 / temp_12[i]))
    metropolis_12_10.append(np.exp(-10 / temp_12[i]))
```

```python
plt.plot(iterations, metropolis_6_1, label = '6 periods, Δ = 1')
plt.plot(iterations, metropolis_6_10, label = '6 periods, Δ = 10')
plt.plot(iterations, metropolis_12_1, label = '12 periods, Δ = 1')
plt.plot(iterations, metropolis_12_10, label = '12 periods, Δ = 10')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Likelihood of acceptance')
plt.tight_layout()
plt.savefig('AcceptancePlot.png')
```
