CDF

```python
def cdf(x):
    if x<0:
        return 0
    if (x>= 0 and x<1):
        return 0.25*(x**2)
    if (x>= 1 and x<2):
        return 0.5*x-0.25
    if (x>= 2 and x<3):
        return (-0.25*x**2+1.5*x-1.25)
    else:
        return 1
```

```python
x = np.linspace(-1,4,100)
y = []
for i in range(len(x)):
   y.append(cdf(x[i]))
```

```python
plt.plot(x,y)
plt.xlabel('t')
plt.ylabel('cdf(t)')
plt.tight_layout()
plt.savefig('Task4_plot_cdf.png')
```

Inverse CDF

```python
def inv_cdf(x):
    if (x>= 0 and x<0.25):
        return 2*math.sqrt(x)
    if (x>= 0.25 and x<0.75):
        return (4*x+1)/2
    if (x>= 0.75 and x<=1):
        return 3-2*math.sqrt(-x+1)
```

```python
x = np.linspace(0,1,100)
y = []
for i in range(len(x)):
   y.append(inv_cdf(x[i]))
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('invcdf(x)')
plt.tight_layout()
plt.savefig('Task4_plot_invcdf.png')
```
